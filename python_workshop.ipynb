````python
def solve(a, b, c):
    delta = (b ** 2) - (4 * a * c)
    x1 = (-b + delta) / 2
    x2 = (-b - delta) / 2
    return x1, x2


ans = solve(1, 1, -2)

print(ans)
print(type(ans))
````

````python
def bubble_sort(l, reversed=False):
    if len(l) <= 1:
        return l
    for i in range(len(l) - 1):
        if (not reversed) and (l[i] > l[i + 1]):
            l[i], l[i + 1] = l[i + 1], l[i]
        elif reversed and (l[i] < l[i + 1]):
            l[i], l[i + 1] = l[i + 1], l[i]
    return bubble_sort(l[:-1], reversed) + [l[-1], ]


l = [1, 6, 3, 8, 4]

s1 = bubble_sort(l)
print(s1)

s2 = bubble_sort(l, reversed=True)
print(s2)

s3 = bubble_sort(l, True)
print(s3)
````

## Class

````python
class ComplexNumber(object):

    def __init__(self, r, i):
        self.r = r
        self.i = i

    def __add__(self, other):
        return ComplexNumber(self.r + other.r, self.i + other.i)

    def __str__(self) -> str:
        return str(self.r) + ' + ' + str(self.i) + 'i'

    def dis(self, other):
        return 1


c1 = ComplexNumber(1, 1)
c2 = ComplexNumber(3, 4)

i = c1.dis(c2)
i = ComplexNumber.dis(c1, c2)

c3 = c1 + c2

print(c3)
````

## Scoping Decipline
for students who are familiar with design of programming languages.

````python
def odd(n: int) -> bool:
    if n == 0:
        return False
    return even(n - 1)


def even(n: int) -> bool:
    if n == 0:
        return True
    return odd(n - 1)


print(odd(100))
````
